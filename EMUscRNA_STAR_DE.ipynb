```r
library(tidyverse)
library(DESeq2)
library(xlsx)
```

```r
input_path <- "./Lysis_scRNAseq_ensembl_NTCs_nonsinglets_removed.csv"
#input_path <- "./temp.csv"

annotate = TRUE
```

```r
make_cntmtx <- function(sample){
    cntmtx = read.table(sample, sep = ",") %>% as_tibble()
    colnames(cntmtx) <- cntmtx[1,]
    cntmtx <- cntmtx[-1,]
    cntmtx[,-1] <- apply(cntmtx[,-1], 2, as.numeric)
    cntmtx
}
```

```r
df <- make_cntmtx(input_path)
df
```

```r
l = list()
for (col in colnames(df)[-1]){
    l <- append(l, filter(sampdata, Sample_ID == str_split(col, "_S")[[1]][2])["Sample_Plate"] %>% as.character())
}
l
```

```r
coldata <- c(unlist(l)) %>% as.data.frame()
rownames(coldata) <- colnames(df)[-1]
colnames(coldata) <- "cell_line"
coldata
```

```r
df <- df %>% as.data.frame()
rownames(df) <- df[,1]
#Use this only if run on data with transcript sums!!
df <- df[-61164,-1]
#df <- df[,-1]
df
```

```r
df = df[,] + 1
```

```r
all(rownames(coldata) == colnames(df))
```

```r
dds <- DESeqDataSetFromMatrix(countData = df,
                              colData = coldata,
                              design = ~ cell_line)
```

```r
dds <- DESeq(dds, test="LRT", reduced=~1, minReplicatesForReplace=Inf, useT=TRUE, minmu=0.1)
res <- results(dds)
res
```

```r
res <- results(dds, contrast=c("cell_line","MCF7","BT474_SETB"))
```

```r
res <- res %>% as.data.frame() %>% arrange(padj)
res
```

```r
if (annotate == TRUE) {
    library(org.Hs.eg.db)
    cols <- c("SYMBOL", "GENENAME")
    genes <- select(org.Hs.eg.db, keys=rownames(res), columns=cols, keytype="ENSEMBL")
    genes <- genes %>% filter(SYMBOL != "NA")
    res["ENSEMBL"] = rownames(res)
    ann_res <- left_join(res, genes,  by="ENSEMBL")
    ann_res
}
```
